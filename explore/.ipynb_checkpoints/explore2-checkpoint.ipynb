```python
import numpy as np
import pandas as pd
import sys,os,json
from collections import Counter

sys.path.insert(1, '/home/chacoma/librerias/python')
from plotter_lib import *

db = '/home/chacoma/Lineas/football_4/data/'
```

```python
arx =db+'Sample_Game_1/X_DS1_raw.json' 

d = json.load(open(arx,'r'))
```

```python
T = len(d["2"])

X1= np.array(d["1"])
X2= np.array(d["2"])

print (T, X1.shape, X2.shape )
```

```python
V = (X1[1] - X1[0])

u = V[0]/np.linalg.norm(V[0])

print (u)
print (V[1]/np.linalg.norm(V[1]))

u+= V[1]/np.linalg.norm(V[1])

print (u)

```

```python
P,P1,P2 = [],[],[]

for t in range(T-1):
    
    V1 = X1[t+1]-X1[t]
    V2 = X2[t+1]-X2[t]
    
    u1=0
    u2=0
    
    N = len(V1)
    
    for n in range(N):
        
        v1norm = np.linalg.norm(V1[n])
        v2norm = np.linalg.norm(V2[n])
        
        if v1norm:
            u1 += V1[n]/v1norm
            
        if v2norm:
            u2 += V2[n]/v2norm
       
            
    P1.append( np.linalg.norm(u1)/(1.0*N) )
    P2.append( np.linalg.norm(u2)/(1.0*N) )
    
    u = u1+u2
    P.append( np.linalg.norm(u)/(2.0*N) )
```

```python
x = np.arange( 0,T-1,1 )


fig, ax = plt.subplots(3,1,figsize=(16,6))

ax[0].plot(x, P1)
ax[1].plot(x, P2, c='C3')
ax[2].plot(x, P, c='black')

t1=1200


ax[0].set_xlim(t1,t1+200)
ax[1].set_xlim(t1,t1+200)
ax[2].set_xlim(t1,t1+200)

plt.show()
```

```python
np.corrcoef(P1,P2)
```
